# Solving a map 

```python
def get_map(id=1, batch='easy'):
    return Map.from_yaml(f"tests/{batch}_map{id}.yaml")
```

```python
from search_methods.beam_search import BeamSearch
from sokoban.moves import *
def solve_map(id=1, batch='easy'):
    map = get_map(id, batch=batch)
    solver = BeamSearch(map)
    map.plot_map()
    state, is_ok = solver.solve()
    print(f"{id=} {batch=}")
    print(is_ok)
    state[1].plot_map()
    print(state[0])
    print("\n\n")
    return state, is_ok
```

# Visualizing a map 

```python
maps = [(1, 'easy'), (2, 'easy'), (1, 'medium'), (2, 'medium'), (1, 'large')]
# maps = [(2, 'hard')]
```

```python
solved = []
for map_params in maps:
    solve_map(*map_params)
```
