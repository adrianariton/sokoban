# Solving a map 

```python
def get_map(id=1, batch='easy'):
    return Map.from_yaml(f"tests/{batch}_map{id}.yaml")
```

```python
from search_methods.beam_search import BeamSearch
from sokoban.moves import *
import time
def solve_map(id=1, batch='easy'):
    map = get_map(id, batch=batch)
    solver = BeamSearch(map, K=20, pulls=True)
    map.plot_map()
    begin_time = time.time()
    state, is_ok = solver.solve()
    end_time = time.time()
    delta_time = end_time - begin_time
    print(f"{id=} {batch=}")
    print(is_ok)
    state[1].plot_map()
    print(state[0])
    print(state[2])
    print("\n\n")
    with open(f'logs/{solver.algo_name}/{map.test_name}.yaml','w+') as f:
        print(f"""moves={state[2]}\nmoves_no={len(state[2])}\npulls={list([m for m in state[2] if m > 4])}\ntime={delta_time}\nexpanded={solver.expanded_states}""", file=f)
    
    return state, is_ok
```

# Visualizing a map 

```python
# maps = [(1, 'easy'), (2, 'easy'), (1, 'medium'), (2, 'medium'), (1, 'large'), (2, 'hard'), (2, 'large')]
maps = [(1, 'super_hard')]
```

```python
solved = []
for map_params in maps:
    solve_map(*map_params)
```
